```python
def load_sentiment_model():
    """
    Initialize and return sentiment analysis model with error handling

    Returns:
        pipeline: Hugging Face sentiment analysis pipeline

    Exits:
        If model fails to load, exits with status 1
    """
    try:
        print("🔄 Loading sentiment analysis model...")
        # Create sentiment analysis pipeline with specific configuration:
        model = pipeline(
            "sentiment-analysis",  # Task type
            model="distilbert-base-uncased-finetuned-sst-2-english",  # Efficient distilled model
            device=-1  # -1 for CPU, 0 for GPU if available
        )
        print("✅ Model loaded successfully!")
        return model
    except Exception as e:
        # Handle any errors during model loading
        print(f"❌ Failed to load model: {str(e)}")
        print("Please check your internet connection and try again.")
        exit(1)  # Exit with error code
```

```python
def get_user_input():

    while True:
        try:

            text = input("\n💬 Enter text to analyze (or 'quit' to exit): ").strip()


            if text.lower() == 'quit':
                exit(0)


            if not text:
                raise ValueError("Input cannot be empty")
            if len(text.split()) > 512:
                raise ValueError("Text too long (max 512 words)")

            return text

        except ValueError as e:

            print(f"⚠️ {str(e)}. Please try again.")
```

```python
def load_sentiment_model():

    try:
        print("🔄 Loading sentiment analysis model...")

        model = pipeline(
            "sentiment-analysis",
            model="distilbert-base-uncased-finetuned-sst-2-english",
            device=-1
        )
        print("✅ Model loaded successfully!")
        return model
    except Exception as e:

        print(f"❌ Failed to load model: {str(e)}")
        print("Please check your internet connection and try again.")
        exit(1)
```

```python
def display_result(text, result):

    print("\n🔍 Sentiment Analysis Result:")
    print("-" * 40)


    print(f"📝 Text: {text[:200] + '...' if len(text) > 200 else text}")


    print(f"🏷️  Sentiment: {result[0]['label'].upper()}")


    print(f"🎯 Confidence: {result[0]['score']:.1%}")

    print("-" * 40)
```

```python
if __name__ == "__main__":

    print("\n" + "=" * 50)
   # print("🌟 TEXT SENTIMENT ANALYZER".center(50))
   # print("=" * 50)

    model = load_sentiment_model()


    while True:
        try:

            text = get_user_input()


            result = model(text)


            display_result(text, result)


            if input("\nAnalyze another? (y/n): ").lower() != 'y':
                print("\n👋 Goodbye!")
                break

        except KeyboardInterrupt:

            print("\n👋 Operation cancelled by user")
            break
        except Exception as e:

            print(f"\n❌ Unexpected error: {str(e)}")
            print("Please try different text or try again later.")
```
